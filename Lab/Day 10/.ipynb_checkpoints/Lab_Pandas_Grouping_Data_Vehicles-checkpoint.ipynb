```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("vehicles.csv")
```

```python
cars.head(10)
```

How many Car models?

group by the data by the Make using count function

```python
cars.groupby('Make').count()
```

What brand has the most cars?

```python
cars['Make'].value_counts().index[0]
```

show the average CO2_Emission_Grams/Km by Brand

```python
cars['CO2 Emission Grams/Km'] = cars['CO2 Emission Grams/Mile']/1.60934
cars.drop(['CO2 Emission Grams/Mile'], axis=1, inplace=True)
```

```python
cars.groupby('Make')[['CO2 Emission Grams/Km']].mean()
```

show the average CO2_Emission_Grams/Km by Brand ... sorted

```python
cars.groupby('Make')[['CO2 Emission Grams/Km']].mean().sort_values('CO2 Emission Grams/Km')
```

Use pd.cut or pd.qcut to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].describe()
```

```python
cars['Decade'] = pd.cut(cars['Year'], 4, labels=['80s','90s','00s','10s'])
```

Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
cars['City Km/Liter'] = cars['City MPG']*(1.60934 / 3.78541)
cars['Highway Km/Liter'] = cars['Highway MPG']*(1.60934 / 3.78541)
cars['Combined Km/Liter'] = cars['Combined MPG']*(1.60934 / 3.78541)
cars.drop(['City MPG', 'Highway MPG', 'Combined MPG'], axis=1, inplace=True)
```

```python
cars
```

```python
cars.groupby('Decade')[['City Km/Liter']].mean()
```

Which brands are more environment friendly?

```python
cars.groupby(['Decade', 'Make'])[['CO2 Emission Grams/Km']].mean()
```

Does the drivetrain affect fuel consumption?

```python
cars.groupby(['Drivetrain'])[['City Km/Liter', 'Highway Km/Liter', 'Combined Km/Liter']].mean()
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
def binary_transmission(x):
    if 'Auto' in x:
        return 'Automatic'
    elif 'Manual' in x:
        return 'Manual'
    else:
        return x
```

```python
cars['Transmission'] = list(map(binary_transmission, cars['Transmission']))
```

```python
cars.groupby(['Transmission'])[['City Km/Liter']].mean()
```

Use groupby and aggregate with different aggregation measures for different columns:

aggregate with average City_Km/Liter and the count of the Trans

```python
cars.groupby(['Transmission']).agg({'City Km/Liter':'mean', 'Transmission': 'count'})
```

aggregate with average City_Km/Liter and the minimum of the Trans

```python
cars.groupby(['Transmission']).agg({'City Km/Liter':'min'})
```
